```python
filenames = glob('Namesbystate\\*.TXT')
```

```python
len(filenames)
```

Merging all the files 

```python
dataframes = []

for names in filenames:
    df = pd.read_csv(names, header=None, names=['State', 'Gender', 'Year', 'Names', 'Count'])
    dataframes.append(df)
```

```python
dataframes
```

```python
len(dataframes)
```

```python
df = pd.concat(dataframes, ignore_index=True)
df
```

```python
df.info()
```

```python
df.to_csv('US_baby_names_by_state.csv', index=False)
```
